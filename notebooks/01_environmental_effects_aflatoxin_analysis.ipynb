```python
def set_boundary(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
```

```python
fig = plt.figure(figsize = (8,9))
#plt.title("Top 10 states with most active cases in India", size = 25)
#iloc is for index location
ax = sns.barplot(data = df1, y = "AspergillusCount", x = "Moisture", linewidth = 2, edgecolor = "red")
plt.xlabel("Moisture")
plt.ylabel("Aspergillus Count (10^6) CFU")
#plt.legend()
plt.savefig("Effect of Moisture on AC_BC.png", format="png", bbox_inches="tight")




plt.show()
```

```python
fig = plt.figure(figsize = (20,10))
ax0 = plt.subplot(111)
ax0.plot(df1.Moisture, df1.AspergillusCount, 'o-', color="black", label='Asperg. Count')
plt.title("Effect of Moisture on Aspergillus Count")
plt.xlabel("Moisture")
plt.ylabel("Aspergillus Count (10^6) CFU")
#plt.legend()
plt.savefig("Effect of Moisture on AC.png", format="png", bbox_inches="tight")




plt.show()
```

```python
fig = plt.figure(figsize = (20,10))
ax1 = plt.subplot(111)
#ax1.plot(df1.Moisture, df1.AspergillusCount, 'o-', color="black", label='Asperg. Count')
ax1.plot(df1.Moisture, df1.AFB1, 'o-', label='AFB1')
ax1.plot(df1.Moisture, df1.AFB2, 's-', label='AFB2')
ax1.plot(df1.Moisture, df1.AFG1, 'x-', label='AFG1')
ax1.plot(df1.Moisture, df1.AFG2, '>-', label='AFG2')
set_boundary(ax1)
ax1.legend(frameon = False)

plt.title("Effect of Moisture on Aflatoxins production by Aspergillus Count")
plt.xlabel("Moisture")
plt.ylabel("Aflatoxin Concentration (ng)")
#plt.legend()
plt.savefig("Effect of Moisture on Aflatoxins.png", format="png", bbox_inches="tight")




plt.show()
```

# Statistical Analysis

### Check normality and significant differences.
## Correlation

```python
from scipy import stats
from scipy.stats import shapiro
from scipy .stats import wilcoxon
```

```python
stat,p=shapiro(df5.AspergillusCount)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
stat,p=shapiro(df5.AFB1)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
#significant difference between AC and AFB1: They are significantly different
wilcoxon(df5.AspergillusCount, df5.AFB1, correction=True)
```

```python
corrr = df5.corr()
corrr
```

### Moisture 4%

```python

stat,p=shapiro(df5.iloc[0:6].AspergillusCount)
print('stat=%.4f, p=%.4f\n'% (stat, p))

#It is a normal distribution
```

```python
# Normal Distribution
stat,p=shapiro(df5.iloc[0:6].AFB1)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
t_stat, p_val = stats.ttest_ind(df5.iloc[0:6].AspergillusCount, df5.iloc[0:6].AFB1)  
print("t-statistic = " + str(t_stat))  
print("p-value = " + str(p_val))
#There is significant difference
```

```python
# not Normal Distribution
stat,p=shapiro(df5.iloc[0:6].AFB2)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
wilcoxon(df5.iloc[0:6].AspergillusCount, df5.iloc[0:6].AFB2, correction=True)
#There is significant difference
```

```python
# not Normal Distribution
stat,p=shapiro(df5.iloc[0:6].AFG1)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
wilcoxon(df5.iloc[0:6].AspergillusCount, df5.iloc[0:6].AFG1, correction=True)
#there is significant difference
```

```python
stat,p=shapiro(df5.iloc[0:6].AFG2)
print('stat=%.4f, p=%.4f\n'% (stat, p))
```

```python
wilcoxon(df5.iloc[0:6].AspergillusCount, df5.iloc[0:6].AFG2, correction=True)
#There is no significant difference
```

```python
t_stat, p_val = stats.ttest_ind(df5.iloc[0:6].AFB1, df5.iloc[0:6].AFG2)  
print("t-statistic = " + str(t_stat))  
print("p-value = " + str(p_val))
#There is significant difference
```
